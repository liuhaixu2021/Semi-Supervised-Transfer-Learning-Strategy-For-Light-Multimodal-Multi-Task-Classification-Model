```python
### deal with the csv loading error
def preprocess_csv(file_path, output_path, istraining=True):
    with open(file_path, 'r', encoding='utf-8') as infile, \
         open(output_path, 'w', encoding='utf-8') as outfile:
        for line in infile:
            line = line.replace('"', ' ')
            first_comma_index = line.find(',')
            if istraining:
              second_comma_index = line.find(',', first_comma_index + 1)
              part_before_second_comma = line[:second_comma_index + 1]
              part_after_second_comma = line[second_comma_index + 1:].replace(',', ' ')
              new_line = part_before_second_comma + part_after_second_comma
            else:
              part_before_first_comma = line[:first_comma_index + 1]
              part_after_first_comma = line[first_comma_index + 1:].replace(',', ' ')
              new_line = part_before_first_comma + part_after_first_comma

            outfile.write(new_line)
### fixed error line
def fixed_csv(df):
  for index, row in df.iterrows():
      # Checks if Labels and Caption are empty in the current row.
      if pd.isna(row['Labels']) and pd.isna(row['Caption']):
          # Check if it's not the last line
          if index + 1 < len(df):
              # Add the ImageID of the next row to the Caption of the current row.
              df.at[index - 1, 'Caption'] = df.at[index - 1, 'Caption']+" "+str(df.at[index, 'ImageID'])
  return df

preprocess_csv('kaggle/input/multi-label-classification-competition-2024/COMP5329S1A2Dataset/train.csv', 'kaggle/input/multi-label-classification-competition-2024/COMP5329S1A2Dataset/fixed_train.csv')
fixed_data = pd.read_csv('kaggle/input/multi-label-classification-competition-2024/COMP5329S1A2Dataset/fixed_train.csv', quotechar='"', escapechar='\\')
fixed_csv(fixed_data)
fixed_data=fixed_data.dropna(subset=['Labels'])
fixed_data.to_csv('fixed_train.csv', index=False)

preprocess_csv('kaggle/input/multi-label-classification-competition-2024/COMP5329S1A2Dataset/test.csv', 'kaggle/input/multi-label-classification-competition-2024/COMP5329S1A2Dataset/fixed_test.csv',istraining=False)
fixed_data = pd.read_csv('kaggle/input/multi-label-classification-competition-2024/COMP5329S1A2Dataset/fixed_test.csv', quotechar='"', escapechar='\\')
fixed_data.to_csv('fixed_test.csv', index=False)
```

```python
nltk.download('wordnet')
nltk.download('omw-1.4')

### random replace words form caption
def synonym_replacement(sentence, n):
    words = sentence.split()
    new_words = words.copy()
    random_word_list = list(set([word for word in words if wordnet.synsets(word)]))
    random.shuffle(random_word_list)
    num_replaced = 0
    for random_word in random_word_list:
        synonyms = set()
        for syn in wordnet.synsets(random_word):
            for lemma in syn.lemmas():
                if lemma.name() != random_word:
                    synonyms.add(lemma.name().replace("_", " ").replace("-", " ").lower())
        if synonyms:
            synonym = random.choice(list(synonyms))
            new_words = [synonym if word == random_word else word for word in new_words]
            num_replaced += 1
        if num_replaced >= n:  # n is the number of words you wish to replace
            break

    sentence = ' '.join(new_words)
    return sentence

data = train_data
### Enhancement of text data
def augmented_data(data):

  augmented_data = []
  for index, row in data.iterrows():
      original_caption = row['Caption']
      augmented_data.append({'ImageID': row['ImageID'], 'Labels': row['Labels'], 'Caption': original_caption})
      for i in range(1):
          new_caption = synonym_replacement(original_caption, 2)
          augmented_data.append({'ImageID': row['ImageID'], 'Labels': row['Labels'], 'Caption': new_caption})


  return pd.DataFrame(augmented_data)

# train_data=augmented_data(train_data)
# train_data.to_csv('augmented_dataset.csv', index=False)
# print(train_data)
# encoded_labels_train,label_index=encoded_labels(train_data)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
tokenizer = BertTokenizer.from_pretrained('google/bert_uncased_L-2_H-128_A-2')
# Dataset class for multi-modal data including texts and images.
class MultiModalDataset(Dataset):
    def __init__(self, texts, image_paths, labels, text_transform, image_transform):
        self.texts = texts
        self.image_paths = image_paths
        self.labels = labels
        self.text_transform = text_transform
        self.image_transform = image_transform

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):

        text = self.texts[idx]
        text_inputs = self.text_transform(text, padding='max_length', max_length=64, truncation=True, return_tensors="pt").input_ids.squeeze(0)

        image = Image.open(self.image_paths[idx])
        if self.image_transform:
            image = self.image_transform(image)

        label = self.labels[idx]

        return image,text_inputs, label
# Dataset class for testing multi-modal data including texts and images.
class MultiModalTestDataset(Dataset):
    def __init__(self, image_paths, texts, transform=None,tokenizer=tokenizer):
        self.image_paths = image_paths
        self.texts = texts
        self.transform = transform
        self.tokenizer = tokenizer
    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, index):
        image = Image.open(self.image_paths[index])
        if self.transform:
            image = self.transform(image)

        text = self.texts[index]
        text_inputs = self.tokenizer(text, padding='max_length', max_length=64, truncation=True, return_tensors="pt").input_ids.squeeze(0)
        return image, text_inputs

# train image transform
image_transform = transforms.Compose([
    transforms.Resize((384, 384)),
    transforms.ToTensor(),
    transforms.RandomHorizontalFlip(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# test image transform
transform = transforms.Compose([
    transforms.Resize((384, 384)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# load trainset
image_dir = 'kaggle/input/multi-label-classification-competition-2024/COMP5329S1A2Dataset/data/'
train_dataset = MultiModalDataset(train_data['Caption'], train_data['ImageID'].map(lambda x: os.path.join(image_dir, x)), encoded_labels_train, tokenizer, image_transform)
#train_loader = DataLoader(train_dataset, batch_size=58, shuffle=True)

# load testset
test_dataset = MultiModalTestDataset(test_data['ImageID'].map(lambda x: os.path.join(image_dir, x)), test_data['Caption'], transform,tokenizer)
test_loader = DataLoader(test_dataset, batch_size=128, shuffle=False)


val_dataset = MultiModalDataset(val_data['Caption'], val_data['ImageID'].map(lambda x: os.path.join(image_dir, x)), encoded_labels_val, tokenizer, image_transform)
val_loader = DataLoader(val_dataset, batch_size=58, shuffle=True)

```

## NLP Model

```python
# Custom BERT model for multi-label sequence classification
class BertForMultiLabelSequenceClassification(nn.Module):
    def __init__(self, num_labels=18,model=BertModel.from_pretrained('google/bert_uncased_L-2_H-128_A-2')):
        super().__init__()
        self.bert = model
        # print(self.bert.config,self.bert)
        config = BertConfig(
            vocab_size=30522,
            hidden_size=256,
            num_hidden_layers=2,
            num_attention_heads=2,
            intermediate_size=1024,
            hidden_act="gelu",
            hidden_dropout_prob=0.1,
            attention_probs_dropout_prob=0.1,
            max_position_embeddings=512,
            type_vocab_size=2,
            initializer_range=0.02,
            layer_norm_eps=1e-12,
            output_attentions=False,
            output_hidden_states=False,
            classifier_dropout=0.1
            )
        # self.bert = BertModel(config)
        self.LN = nn.LayerNorm(self.bert.config.hidden_size)
        self.classifier = nn.Linear(self.bert.config.hidden_size, num_labels)

    def forward(self, text_inputs, images=None,attention_mask=None):
        outputs = self.bert(text_inputs, attention_mask=attention_mask).pooler_output
        outputs=self.LN(outputs)
        logits = self.classifier(outputs)
        return logits


# modelbert = BertForMultiLabelSequenceClassification()
# bertmodel=modelbert.bert
```

## Fusion model

```python
# Custom EfficientNet model for feature extraction
class CustomEfficientNet(nn.Module):
    def __init__(self, model):
        super(CustomEfficientNet, self).__init__()
        self.features = nn.Sequential(*list(model.children())[:-1])

    def forward(self, images):
        x = self.features(images)
        return x

# Cross attention mechanism for combining text and image features
class CrossAttention(nn.Module):
    def __init__(self, text_dim,image_dim,hidden_dim):
        super(CrossAttention, self).__init__()
        self.query_proj = nn.Linear(text_dim, hidden_dim)
        self.key_proj = nn.Linear(image_dim, hidden_dim)
        self.value_proj = nn.Linear(image_dim, hidden_dim)
        self.softmax = nn.Softmax(dim=-1)
        self.linear = nn.Linear(hidden_dim, hidden_dim)

    def forward(self, query, key, value):
        queries = self.query_proj(query)
        keys = self.key_proj(key)
        values = self.value_proj(value)

        attn_scores = torch.matmul(queries, keys.transpose(-2, -1))
        attn_probs = self.softmax(attn_scores)

        weighted_values = torch.matmul(attn_probs, values)
        output = self.linear(weighted_values)
        return output
# Fusion model combining text and image features with cross attention and residual layer
class FusionModalModel(nn.Module):
    def __init__(self,bertmodel,original_model,num_labels=18,cutoff=True,hidden_dim=128,num_classes=18):
        super().__init__()
        self.cutoff=cutoff
        self.hidden_dim=hidden_dim
        self.text_model=bertmodel
        self.image_model = CustomEfficientNet(original_model)
        if self.cutoff:
            self.concat_layer = nn.Linear(self.text_model.config.hidden_size + 1792, hidden_dim)
            self.cross_attention = CrossAttention(self.text_model.config.hidden_size, 1792, hidden_dim)
            self.classifier = nn.Linear(hidden_dim, num_classes)
        else:
            self.cross_attention = CrossAttention(self.text_model.config.hidden_size, 1792, self.text_model.config.hidden_size + 1792)
            self.classifier = nn.Linear(hidden_dim, num_classes)

    def forward(self, text_inputs, images):
        text_outputs = self.text_model(text_inputs).pooler_output
        image_outputs = self.image_model(images)
        combined_features = torch.cat((text_outputs, image_outputs), dim=1)
        if self.cutoff:
          cutoff = self.concat_layer(combined_features)
          fused_features = self.cross_attention(text_outputs, image_outputs, image_outputs)
          logits = self.classifier(fused_features+cutoff)
          return logits
        else:
          cutoff=combined_features
          fused_features = self.cross_attention(text_outputs, image_outputs, image_outputs)
          logits = self.classifier(fused_features+cutoff)
          return logits
# Multi-modal model combining text and image features for classification
class MultiModalModel(nn.Module):
    def __init__(self,bertmodel,original_model,num_labels=18):
        super().__init__()
        self.text_model=bertmodel
        self.image_model = CustomEfficientNet(original_model)
        #self.classifier = nn.Linear(self.text_model.config.hidden_size + self.image_model.classifier.in_features, num_labels)
        self.classifier = nn.Linear(self.text_model.config.hidden_size + 1792, num_labels)
    def forward(self, text_inputs, images):
        text_outputs = self.text_model(text_inputs).pooler_output
        image_outputs = self.image_model(images)
        # combined features
        combined_features = torch.cat((text_outputs, image_outputs), dim=1)
        logits = self.classifier(combined_features)
        return logits

```

## Training

```python
from json import encoder
class Trainer:
  def __init__(self,model,train_dataset,val_dataset):
    super().__init__()
    self.model = model
    self.train_dataset = train_dataset
    self.val_dataset = val_dataset


  def train(self,epochs=6,lr=5e-4,weights_tensor=weights_tensor,batch_size=58,name="",optimizer="adam"):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    train_loader = DataLoader(self.train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(self.val_dataset, batch_size=96, shuffle=True)
    model=self.model.to(device)
    criterion = nn.BCEWithLogitsLoss(pos_weight=weights_tensor.to(device))
    if optimizer=="adam":
      optimizer = optim.Adam(model.parameters(), lr=lr)
    elif optimizer=="SGD":
      optimizer = optim.SGD(model.parameters(), lr=lr)
    else:
      optimizer = optim.AdamW(model.parameters(), lr=lr)

    for epoch in range(epochs):  # adjust as necessary for the number of epochs
      model.train()
      total_loss = 0
      progress_bar=tqdm(train_loader, desc=f"Epoch {epoch+1}")
      for  images,text_inputs, labels in  progress_bar:
          text_inputs, images, labels = text_inputs.to(device), images.to(device), labels.to(device)
          optimizer.zero_grad()
          outputs = model(text_inputs=text_inputs, images=images)
          loss = criterion(outputs, labels.float())
          loss.backward()
          optimizer.step()
          total_loss += loss.item()
          progress_bar.set_postfix(loss=loss.item())

      average_loss = total_loss / len(train_loader)
      val_loss, val_f1,time = self.validate(model, val_loader, criterion, device)  # Validate after each epoch
      print(f'Epoch {epoch+1}, Training Loss: {average_loss}, Validation Loss: {val_loss}, Validation F1 Score: {val_f1}')

    torch.save(model.state_dict(), f'best_model_{name}.pth')
    return model


  def validate(self, model,data_loader, criterion, device):
    import time
    start_time = time.time()
    model.eval()
    total_val_loss = 0
    all_labels = []
    all_preds = []
    with torch.no_grad():
        progress_bar=tqdm(data_loader)
        for images, text_inputs, labels in progress_bar:
            text_inputs, images, labels = text_inputs.to(device), images.to(device), labels.to(device)
            outputs = model(text_inputs=text_inputs, images=images)
            loss = criterion(outputs, labels.float())
            total_val_loss += loss.item()

            predictions = torch.sigmoid(outputs).cpu().numpy() > 0.5
            all_preds.extend(predictions)
            all_labels.extend(labels.cpu().numpy())
    end_time = time.time()
    time=end_time-start_time
    print(time)

    average_val_loss = total_val_loss / len(data_loader)
    f1 = f1_score(all_labels, all_preds, average='micro')

    return average_val_loss, f1,time


  def predict(self,test_dataset,test_data,method,modelbert=None,efficientmodel=None,model=None,plot=False,label_index=label_index):
    predictions = []
    test_loader = DataLoader(test_dataset, batch_size=128, shuffle=False)
    with torch.no_grad():
        for data in tqdm(test_loader, total=len(test_loader), desc='Predicting'):
            images, texts = data[0], data[1]
            images = images.to('cuda')
            #text_inputs = tokenizer(texts, padding='max_length', max_length=max_length, truncation=True, return_tensors="pt").input_ids
            text_inputs = texts.to('cuda')
            if modelbert is not None:
              outputs_text=modelbert(text_inputs=text_inputs)
              predicted_text = torch.sigmoid(outputs_text).cpu().numpy()

            if efficientmodel is not None:
              outputs_image = efficientmodel(images=images)
              predicted_image = torch.sigmoid(outputs_image).cpu().numpy()

            if model is not None:
              predicted_fusion=model(text_inputs,images)
              predicted_fusion = torch.sigmoid(predicted_fusion).cpu().numpy()

            if method=='fusion':
              predicted = predicted_fusion
            elif method=='text':
              predicted = predicted_text
            elif method=='image':
              predicted = predicted_image
            elif method=='logits_fusion':
              predicted = (predicted_text+predicted_image)/2
            elif method=='fully_fusion':
              predicted =(predicted_text+predicted_image+predicted_fusion)/3

            predictions.extend(predicted)
    predictions = np.array(predictions)

    results = pd.DataFrame(predictions, columns=[f'Label_{i}' for i in range(predictions.shape[1])])
    results['ImageID'] = test_data['ImageID']
    result=[]
    for i, row in results.iterrows():
        indices = np.where(row[:-1] > 0.5)[0]
        original_labels = [key for key, value in label_index.items() if value in indices]
        if not original_labels:
            max_index = np.argmax(row[:-1])
            max_label = [key for key, value in label_index.items() if value == max_index]
            original_labels.extend(max_label)
        result.append(original_labels)

    results['Labels'] = result
    results["Labels"] = results["Labels"].apply(lambda x: ' '.join(x))
    results=results.drop([f'Label_{i}' for i in range(predictions.shape[1])], axis=1)

    if plot==True:
      encoded_labels_test,label_index=encoded_labels(test_data)
      encoded_labels_submission,label_index=encoded_labels(results)
      score_f1=f1_score(encoded_labels_test,encoded_labels_submission,average='micro')
      score_acc=accuracy_score(encoded_labels_test,encoded_labels_submission)
      score_class=f1_score(encoded_labels_test,encoded_labels_submission,average=None)
      print(f'F1 Score: {score_f1}, Accuracy: {score_acc}, Class F1 Score: {score_class}')
      # Create a list of category label
      categories = [f'Label_{i}' for i in range(predictions.shape[1])]
      plt.figure(figsize=(15, 6))
      plt.bar(categories, score_class, color='blue')
      plt.title('F1 Score by Category')
      plt.xlabel('Categories')
      plt.ylabel('F1 Score')
      plt.show()

    else:
      results.to_csv('submission.csv', index=False)
      return results

```

```python
### new train_data with
def load_train_data():
  submission = pd.read_csv('submission.csv')
  train_data = pd.read_csv('fixed_train.csv')
  test_data = pd.read_csv('fixed_test.csv')
  test_data["Labels"]=submission['Labels']
  result = pd.concat([train_data, test_data], axis=0,ignore_index=True)
  encoded_labels_train,label_index=encoded_labels(result)
  print(encoded_labels_train)
  train_data=result
  print(train_data)
  train_data.to_csv('train.csv', index=False)
  return train_data,encoded_labels_train
train_data,encoded_labels_train=load_train_data()


####reload data
def load_data(train_data,encoded_labels_train):
  image_dir = 'kaggle/input/multi-label-classification-competition-2024/COMP5329S1A2Dataset/data/'
  train_dataset = MultiModalDataset(train_data['Caption'], train_data['ImageID'].map(lambda x: os.path.join(image_dir, x)), encoded_labels_train, tokenizer, image_transform)
  test_dataset = MultiModalTestDataset(test_data['ImageID'].map(lambda x: os.path.join(image_dir, x)), test_data['Caption'], transform,tokenizer)
  # test_loader = DataLoader(test_dataset, batch_size=128, shuffle=False)

  val_dataset = MultiModalDataset(val_data['Caption'], val_data['ImageID'].map(lambda x: os.path.join(image_dir, x)), encoded_labels_val, tokenizer, image_transform)
  ###
  return train_dataset,val_dataset,test_dataset,test_loader

####re-trainning
def semi_train():
  efficientmodel = EfficientNet()
  original_model = efficientmodel.model
  trainer=Trainer(efficientmodel,train_dataset,val_dataset)
  efficientmodel=trainer.train(epochs=2,lr=2e-4,weights_tensor=weights_tensor,batch_size=60,name="efficientnet")
  trainer.predict(test_dataset,test_data,'image',efficientmodel=efficientmodel)
  efficientmodel.to('cpu')

  modelbert = BertForMultiLabelSequenceClassification()
  bertmodel=modelbert.bert
  trainer=Trainer(modelbert,train_dataset,val_dataset)
  modelbert=trainer.train(epochs=3,lr=5e-4,weights_tensor=weights_tensor,batch_size=256,name="bert")
  trainer.predict(test_dataset,test_data,'text',modelbert=modelbert)
  modelbert.to('cpu')

  efficientmodel.to('cuda')
  modelbert.to('cuda')
  trainer.predict(test_dataset,test_data,'logits_fusion',efficientmodel=efficientmodel,modelbert=modelbert)
  trainer.predict(train_dataset,train_data,'logits_fusion',efficientmodel=efficientmodel,modelbert=modelbert,plot=True)
  efficientmodel.to('cpu')
  modelbert.to('cpu')
####Encapsulation with loops
macro_epoches=5
for i in range(macro_epoches):
  train_data,encoded_labels_train=load_train_data()
  train_dataset,val_dataset,test_dataset,test_loader=load_data(train_data,encoded_labels_train)
  semi_train()
```
